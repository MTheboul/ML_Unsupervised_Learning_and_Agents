```python
import numpy as np
import matplotlib.pyplot as plt
```

___

### Part 1

```python
mean_height = 175
std_dev_height = 7
mean_weight = 77
std_dev_weight = 5

z_value = (mean_height, mean_weight)

print("Valeur commune de Z (Taille, Poids) des hommes en France:", z_value)
```

Sachant que Z = (X, Y) avec X le poids moyen des hommes en France en fonction de Y la taille moyenne des hommes en France, X et Y suivent une distribution normale avec deux moyennes, on obtient bien deux valeurs finies.

> _La moyennes et l’écart-type utilisés dans le code ont été récupérer des données postées sur l’INSEE._
___

### Part 2

```python
n = 1000
height_samples = np.random.normal(mean_height, std_dev_height, 100)
weight_samples = np.random.normal(mean_weight, std_dev_weight, 100)

height_samples_euc = np.random.normal(mean_height, std_dev_height, 1000)
weight_samples_euc = np.random.normal(mean_weight, std_dev_weight, 1000)

empirical_averages = []

plt.figure()
plt.scatter(height_samples, weight_samples, alpha=0.5)
plt.xlabel('Taille (cm)')
plt.ylabel('Poids (kg)')
plt.title("Représentation de Z = (Taille, Poids)")
plt.grid(True)
plt.show()
```

Ce premier graphique représente une génération d’échantillon composé de 100 personnes avec leur taille (en cm) et leur poids (en kg).
___

### Part 3

```python
for i in range(1, n+1):
   empirical_average_height = np.mean(height_samples_euc[:i])
   empirical_average_weight = np.mean(weight_samples_euc[:i])
   empirical_average_XY = (empirical_average_height, empirical_average_weight)
   empirical_averages.append(empirical_average_XY)

euclidean_distances = [np.linalg.norm(np.array(empirical_average) - np.array(z_value)) for empirical_average in empirical_averages]

plt.figure()
plt.plot(range(1, n+1), euclidean_distances)
plt.xlabel("Nombre d'échantillon (n)")
plt.ylabel('Distance euclidienne')
plt.title('Convergence de la moyenne empirique à la valeur commune')
plt.grid(True)
plt.show()
```

Le schéma suivant représente la moyenne empirique en fonction de la loi de Z, on constate que la distance euclidienne diminue donc nous pouvons en conclure que les valeurs convergent bien vers la valeur commune lorsque le nombre d'échantillons augmente.
